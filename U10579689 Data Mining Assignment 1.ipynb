```python
res = df.groupby('title').aggregate({'dislikes':'sum','likes':'sum',})
res.sort_values(by=['likes'],ascending=False)
```

```python
most_liked_videos=df.groupby('likes')
most_liked_videos.first().tail(5)
```

```python
max_views=df['views'].max()
print(max_views)
```
